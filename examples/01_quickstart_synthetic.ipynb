# scEvoNet quickstart (synthetic data)

Self-contained example: **random** cells × genes matrix and three toy labels. No downloads.

## Dependencies

```bash
pip install scevonet
```

````python
import warnings

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore", category=UserWarning)

from scevonet import EvoManager, Sample, SampleConfig
````

````python
rng = np.random.default_rng(0)

n_cells, n_genes = 80, 40
genes = [f"G{i}" for i in range(n_genes)]
X = rng.random((n_cells, n_genes)) * 2.0
matrix = pd.DataFrame(X, columns=genes)
labels = ["Neuron"] * 25 + ["Mesenchyme"] * 25 + ["Epi"] * 30
````

````python
cfg = SampleConfig(
    top_features_limit=25,
    n_estimators=60,
    early_stopping_rounds=5,
    random_state=42,
)

sample_a = Sample(matrix, labels, config=cfg)
sample_b = Sample(matrix, labels, config=cfg)

print(f"Cell types modeled: {len(sample_a.models)}")
````

````python
em = EvoManager(sample_a, sample_b, network_top_genes=15)

print("Prediction blocks:", sorted(em.predictions.keys()))
print("Similarity table:", em.cell_types_similarity.shape)
print(em.network.head())
````
